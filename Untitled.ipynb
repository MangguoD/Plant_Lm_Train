```python
import torch
import torch.serialization
from plant_lm_train import DNAModel, Config
torch.serialization.add_safe_globals([Config])
torch.serialization.add_safe_globals([DNAModel])
# 指定你的.pth文件路径
file_path = 'downstream/model/bailizhong_classifier.pth'

try:
    # 加载模型文件，map_location='cpu'确保即使在无GPU环境下也能加载
    checkpoint = torch.load(file_path, map_location='cpu')
    
    # 打印所有键，方便确认数据结构
    print("检查点文件中的键:", checkpoint.keys())
    
    # 尝试获取 best_val_loss 字段
    if 'best_val_loss' in checkpoint:
        best_val_loss_value = checkpoint['best_val_loss']
        print(f"找到 best_val_loss: {best_val_loss_value}")
    else:
        print("该检查点文件中未找到 'best_val_loss' 字段。")
        # 打印所有键和其对应数据的类型，帮助进一步判断
        print("\n文件内容详情:")
        for key, value in checkpoint.items():
            print(f"  {key}: {type(value)}")
            
except FileNotFoundError:
    print(f"错误：在路径 '{file_path}' 下未找到文件。请检查文件路径是否正确。")
except Exception as e:
    print(f"读取文件时发生错误: {e}")
```
